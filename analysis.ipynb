```python
import pandas as pd
```

```python
df=pd.read_csv("marketing_campaign.csv",sep=";")
```

```python
print(df.head())
print(df.shape)
print(df.isnull().sum())
```

```python
df.dropna(inplace=True)
```

```python
df["Age"] = 2024 - df["Year_Birth"]

df["Total_Spend"] = (df["MntWines"] + df["MntFruits"] + 
                     df["MntMeatProducts"] + df["MntFishProducts"] + 
                     df["MntSweetProducts"] + df["MntGoldProds"])

df["Total_Campaigns_Accepted"] = (df["AcceptedCmp1"] + df["AcceptedCmp2"] + 
                                   df["AcceptedCmp3"] + df["AcceptedCmp4"] + 
                                   df["AcceptedCmp5"])
```

```python
df.to_csv("cleaned_campaign.csv", index=False)
print("Done! Cleaned data saved.")
```
